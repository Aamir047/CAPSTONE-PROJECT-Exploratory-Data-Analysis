## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Step 1: I have loaded the dataset
df = pd.read_csv('/Airbnb NYC 2019.csv')

# Step 2: Checked and removed the duplicate rows
df = df.drop_duplicates()

# Step 3: Handled the missing values
# For numerical columns, filled in the missing values with the median
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())

# For categorical columns, I have filled in the missing values with the mode
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

# Step 4: I have converted the columns to appropriate data types
# Converted 'last_review' to datetime format
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Step 5: Removed unnecessary columns
# Example: Dropping a column like 'host_name' that may not be essential for analysis
df = df.drop(columns=['host_name'])

# Step 6: I have reset the index after cleaning
df.reset_index(drop=True, inplace=True)

# Step 7: Checked the dataset info after cleaning
print(df.info())

# Optional: Checked for unique values in each column
unique_values = df.nunique()
print("Unique Values for Each Column:")
print(unique_values)

# Step 8: Quick summary statistics
print("Dataset Summary Statistics:")
print(df.describe())
````

### What all manipulations have you done and insights you found?

**Removed Duplicates: Eliminated any repeated rows.
Handled Missing Values: Filled missing numerical values with the median and categorical values with the mode.
Converted Data Types: Changed last_review to a datetime format.
Dropped Unnecessary Columns: Removed columns like host_name.
Reset Index: Cleaned the dataset index.
Insights:
The dataset is now free from duplicates and missing values.
last_review is now in the correct format for time analysis.
The data is ready for analysis on pricing, reviews, and availability trends.**

#### Chart - 1

````python
# Chart - 1 visualization code
plt.figure(figsize=(10, 6))

# Created a boxplot to visualize the distribution of prices across room types
sns.boxplot(x='room_type', y='price', data=df)

# Added a title and labels
plt.title('Price Distribution Across Different Room Types', fontsize=16)
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)

# Display the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a boxplot to visualize the distribution of prices across different room types because it effectively highlights the central tendency (median), spread (quartiles), and outliers of the data. This allows for a clear comparison of how prices vary between room types, helping identify trends and price differences, such as whether certain room types have higher or more variable prices.

##### 2. What is/are the insight(s) found from the chart?

The boxplot reveals that Entire home/apt listings have significantly higher prices compared to Private rooms and Shared rooms, with a wider range of prices and more outliers. Private rooms tend to have a more consistent price range, while Shared rooms are generally the least expensive, with a lower median price and fewer outliers. This indicates that room type is a major factor influencing pricing on Airbnb.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

The insights gained from the boxplot can indeed create a positive business impact by guiding strategic pricing and marketing decisions. For example, understanding that Entire home/apt listings command higher prices can inform hosts and property managers to better tailor their offerings and pricing strategies for premium listings. Similarly, knowing that Private rooms have more consistent pricing allows for targeted promotions or adjustments in less expensive segments.
However, there could be a potential negative impact if Shared rooms are overly priced or not differentiated enough from private rooms, leading to lower occupancy rates.
